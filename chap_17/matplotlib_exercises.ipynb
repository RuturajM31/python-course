```python
import matplotlib.pyplot as plt
```

```python
# Plot temperature changes over a week.
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temps = [22, 24, 19, 23, 25, 28, 26]

plt.plot(days, temps)
plt.title("Weekly Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.show()
```

1. 📈 Step Chart (Change Over Time)
🧠 Task:

Show data changes in steps (e.g., stock price updates)

```python

days = [1, 2, 3, 4, 5]
price = [100, 102, 101, 105, 107]

plt.step(days, price, where='mid')
plt.title("Stock Price (Step Chart)")
plt.xlabel("Day")
plt.ylabel("Price")
plt.show()
```

2. 📊 Stacked Bar Chart
🧠 Task:

Compare two categories stacked together

```python
labels = ["A", "B", "C"]
x = [1, 2, 3]
y1 = [10, 20, 15]
y2 = [5, 10, 8]

plt.bar(x, y1, label="Product 1")
plt.bar(x, y2, bottom=y1, label="Product 2")

plt.xticks(x, labels)
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

3. 📦 Box Plot (Data Distribution)
🧠 Task:

Show spread of exam scores

```python
import numpy as np
import matplotlib.pyplot as plt

data = [np.random.normal(60, 10, 100),
        np.random.normal(70, 15, 100),
        np.random.normal(80, 5, 100)]

plt.boxplot(data, labels=["Class A", "Class B", "Class C"])
plt.title("Box Plot - Score Distribution")
plt.show()
```

4. 🎻 Violin Plot (Advanced Distribution View)
🧠 Task:

Compare distributions more smoothly

```python
import numpy as np
import matplotlib.pyplot as plt

data = [np.random.normal(50, 10, 100),
        np.random.normal(60, 20, 100)]

plt.violinplot(data)
plt.title("Violin Plot")
plt.show()
```

5. 🔥 Heatmap (Correlation-style Visualization)
🧠 Task:

Visualize a matrix of values

```python
import numpy as np
import matplotlib.pyplot as plt

matrix = np.random.rand(5, 5)

plt.imshow(matrix, cmap="viridis")
plt.colorbar()
plt.title("Heatmap Example")
plt.show()
```
