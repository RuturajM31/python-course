📊 Mini Project: Marketing KPI Dashboard using NumPy
🎯 Goal

We simulate marketing campaign data and calculate key KPIs like:

Total impressions
Clicks
CTR (Click-Through Rate)
Conversion rate
Revenue
Cost & ROI

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# 10 marketing campaigns
campaigns = np.array([
    "Google Ads", "Facebook Ads", "Instagram Ads", "LinkedIn Ads",
    "Email Campaign", "YouTube Ads", "SEO", "Affiliate", "TikTok Ads", "Display Ads"
])

# Simulated data
impressions = np.random.randint(10000, 100000, size=10)
clicks = np.random.randint(500, 10000, size=10)
conversions = np.random.randint(50, 2000, size=10)

cost = np.random.randint(500, 10000, size=10)  # ad spend in €
revenue = np.random.randint(1000, 20000, size=10)
```

📌 Step 2: Basic KPI Calculations (NumPy Operations)

```python
ctr = (clicks / impressions) * 100
cr = (conversions / clicks) * 100
roi = ((revenue - cost) / cost) * 100
```


📊 Step 3: Combine Everything into a Dataset


np.column_stack() is a NumPy function used to combine multiple 1D arrays into a 2D table (matrix) by placing them as columns side by side.


📌 Example
import numpy as np
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])
result = np.column_stack((a, b))
print(result)

Output:
[[ 1 10] [ 2 20] [ 3 30]]

```python
data = np.column_stack([
    impressions,
    clicks,
    conversions,
    cost,
    revenue,
    ctr,
    cr,
    roi
])
```

🧾 Step 4: Show Clean Output Table

You have a list:

campaigns = ["Google Ads", "Facebook Ads", "SEO"]

Python internally numbers them:

Index (i)	Value
0	Google Ads
1	Facebook Ads
2	SEO

print(f"CTR (%) : {ctr[i]:.2f}")

👉 .2f is a formatting rule for numbers in Python.

```python
print("\nMarketing KPI Dashboard:\n")

for i in range(len(campaigns)):
    print(f"Campaign: {campaigns[i]}")
    print(f"  Impressions : {impressions[i]}")
    print(f"  Clicks      : {clicks[i]}")
    print(f"  Conversions : {conversions[i]}")
    print(f"  Cost (€)    : {cost[i]}")
    print(f"  Revenue (€) : {revenue[i]}")
    print(f"  CTR (%)     : {ctr[i]:.2f}")
    print(f"  CR (%)      : {cr[i]:.2f}")
    print(f"  ROI (%)     : {roi[i]:.2f}")
    print("-" * 40)
```

🔍 Step 5: NumPy Filtering (Best Campaigns)

```python
# Best CTR campaigns

best_ctr = campaigns[ctr > 10]
print("High CTR Campaigns:", best_ctr)
```

```python
# Best ROI campaigns

best_roi = campaigns[roi > 50]
print("High ROI Campaigns:", best_roi)
```

```python
# Average performance

print("Avg CTR:", np.mean(ctr))
print("Avg ROI:", np.mean(roi))
print("Avg Conversion Rate:", np.mean(cr))
```

```python
# Best & worst campaign

best_campaign = campaigns[np.argmax(roi)]
worst_campaign = campaigns[np.argmin(roi)]

print("Best Campaign:", best_campaign)
print("Worst Campaign:", worst_campaign)

# 👉 argmax() and argmin() return index positions
```

⚡ Step 7: Broadcasting Example

Suppose we apply a 10% cost increase across all campaigns:
new_cost = cost * 1.10

```python
new_cost = cost * 1.10
```

🧠 Key NumPy Concepts Used
Concept	Where used
Arrays	campaign data storage
Vectorization	KPI calculations
Broadcasting	cost adjustment
Boolean indexing	filtering campaigns
argmax/argmin	best/worst campaigns
statistics	mean CTR, ROI
column_stack	dataset creation

```python

# -----------------------------
# DASHBOARD STYLE FIGURE
# -----------------------------
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("📊 Marketing KPI Dashboard (NumPy Project)", fontsize=16)

# -----------------------------
# 1. ROI BAR CHART
# -----------------------------
axes[0, 0].bar(campaigns, roi)
axes[0, 0].set_title("ROI by Campaign")
axes[0, 0].set_ylabel("ROI (%)")
axes[0, 0].tick_params(axis='x', rotation=45)

# -----------------------------
# 2. CTR BAR CHART
# -----------------------------
axes[0, 1].bar(campaigns, ctr)
axes[0, 1].set_title("CTR by Campaign")
axes[0, 1].set_ylabel("CTR (%)")
axes[0, 1].tick_params(axis='x', rotation=45)

# -----------------------------
# 3. CTR vs Conversions (Scatter)
# -----------------------------
axes[1, 0].scatter(ctr, conversions)
axes[1, 0].set_title("CTR vs Conversions")
axes[1, 0].set_xlabel("CTR (%)")
axes[1, 0].set_ylabel("Conversions")

# Add labels for better readability
for i, txt in enumerate(campaigns):
    axes[1, 0].annotate(txt, (ctr[i], conversions[i]), fontsize=8)
    
# -----------------------------
# 4. Impressions vs Clicks
# -----------------------------
axes[1, 1].plot(campaigns, impressions, marker='o', label="Impressions")
axes[1, 1].plot(campaigns, clicks, marker='o', label="Clicks")
axes[1, 1].set_title("Impressions vs Clicks")
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].legend()


# -----------------------------
# FINAL LAYOUT FIX
# -----------------------------
plt.tight_layout()
plt.show()
```
