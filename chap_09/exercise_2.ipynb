```python
### Exercise 2 — Day of Week Counter

from pathlib import Path

# Ask user for file name
fname = input("Enter file name: ")

# If user just presses Enter, use default file
if len(fname) < 1:
    fname = "mbox-short.txt"

# Build correct path (chap_09 → parent → files folder)
BASE_DIR = Path.cwd().parent
file_path = BASE_DIR / "files" / fname

# Open the file for reading
handle = open(file_path)

# Dictionary to store counts of each weekday
counts = {}

# Loop through each line in the file
for line in handle:

    # We only care about lines that start with "From "
    if line.startswith("From "):

        # Split the line into words
        words = line.split()

        # Safety check (avoid index errors)
        if len(words) > 2:

            # The day of the week is the 3rd word (index 2)
            day = words[2]

            # Increase count for this day
            counts[day] = counts.get(day, 0) + 1

# Print final dictionary (days and counts)
print(counts)
```
