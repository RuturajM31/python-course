```python
# Exercise 4 — Find Most Active Email Sender

from pathlib import Path

# Ask user for file name
fname = input("Enter file name: ")

# If user just presses Enter, use default file
if len(fname) < 1:
    fname = "mbox-short.txt"

# Build correct path (chap_09 → parent → files folder)
BASE_DIR = Path.cwd().parent
file_path = BASE_DIR / "files" / fname

# Open the file for reading
handle = open(file_path)

counts = {}

# Step 1: Build histogram (same as Exercise 3)
for line in handle:

    if line.startswith("From "):

        words = line.split()
        email = words[1]

        counts[email] = counts.get(email, 0) + 1

# Step 2: Find the email with the highest count

big_email = None   # stores email with max messages
big_count = None   # stores max number of messages

# Loop through dictionary items (email, count pairs)
for email, count in counts.items():

    # If first item OR current count is bigger than max so far
    if big_count is None or count > big_count:

        # update max values
        big_email = email
        big_count = count

# Print result: most active email and its count
print(big_email, big_count)
```
