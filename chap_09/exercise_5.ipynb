```python
# Count Domains

from pathlib import Path

# Ask user for file name
fname = input("Enter file name: ")

# If user just presses Enter, use default file
if len(fname) < 1:
    fname = "mbox-short.txt"

# Build correct path (chap_09 → parent → files folder)
BASE_DIR = Path.cwd().parent
file_path = BASE_DIR / "files" / fname

# Open the file for reading
handle = open(file_path)

# Dictionary to store domain → count
counts = {}

for line in handle:

    if line.startswith("From "):

        words = line.split()

        # Extract full email
        email = words[1]

        # Split email into name and domain
        # Example: user@umich.edu → ["user", "umich.edu"]
        domain = email.split("@")[1]

        # Count domain occurrences
        counts[domain] = counts.get(domain, 0) + 1

# Print final domain histogram
print(counts)
```
