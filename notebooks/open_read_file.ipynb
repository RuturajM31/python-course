```python
### Program calculates  word frequecy in given text

import string
from pathlib import Path

BASE_DIR = Path.cwd().parent                      # move from /notebooks → project root
FILES_DIR = BASE_DIR / "files"                    # folder containing text files

name = input("Enter file name (without .txt): ").strip()   # user input

if not name.endswith(".txt"):                     # ensure correct file extension
    name += ".txt"

file_path = FILES_DIR / name                      # full file path inside /files

print("Searching:", file_path)                    # debug: show resolved path

try:

    if not file_path.exists():                    # check if file exists
        raise FileNotFoundError

    counts = {}                                  # word frequency dictionary

    with open(file_path, "r", encoding="utf-8") as handle:

        for line in handle:                      # read file line by line

            for word in line.split():           # split into words

                word = word.lower().strip(string.punctuation)   # normalize word

                if word:                         # ignore empty values
                    counts[word] = counts.get(word, 0) + 1   # count frequency

    bigword, bigcount = None, 0                  # track most frequent word

    for word, count in counts.items():          # iterate through dictionary

        if bigword is None or count > bigcount: # update max if higher found
            bigword, bigcount = word, count

    print("\nResult:")                           # final output section
    print(bigword, bigcount)

except FileNotFoundError:                        # file missing case
    print("File not found in /files folder")

except Exception as e:                           # unexpected errors
    print("Error:", e)
```
